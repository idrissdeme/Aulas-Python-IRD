```python
import matplotlib.pyplot as plt
from math import *

# Usando um Loop while para criar a lista (tabela) dos valores a plotar
valores_ft=[]
valorest=[]
tfinal = 10
tinicial=-10
Total=1000
t = tinicial
Passo= (tfinal-tinicial)/Total

while t < tfinal:
  ft = 3*t+2
  t = t+Passo
  valorest.append(t)
  valores_ft.append(ft)
  
print(valorest)
print(valores_ft)

#Plotando
plt.plot (valorest,valores_ft,'green')
plt.title ('Função de primeiro grau')
plt.xlabel('x(m)')
plt.ylabel('t(s)')
plt.show()

# Exercíco:
#Use este código para plotar a gráfico do decaimento do Césio 137 em função do tempo num periodo de 1000 anos
```

```python
from math import *



def truncate(f, n):
    '''Truncates/pads a float f to n decimal places without rounding'''
    s = '{}'.format(f)
    if 'e' in s or 'E' in s:
        return '{0:.{1}f}'.format(f, n)
    i, p, d = s.partition('.')
    return '.'.join([i, (d+'0'*n)[:n]])

a = 0.5*exp(-10)
print(round(a/exp(-10),4))
print("%.2f" % a)
print(round(a,2))
print(f'{a:.2f}')
print(truncate(a,2))
```
