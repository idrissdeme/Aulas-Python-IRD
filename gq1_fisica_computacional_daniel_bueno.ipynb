```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-talk')
```

```python
# estima o valor de PI usando método de Monte Carlo.
def pi(n=100, plot=False):
  # matriz 2xn de pontos x e y.
  points = np.random.rand(n, 2)
  # pontos x e y no interior do circulo
  result = np.sqrt(points[:, 0]**2 + points[:, 1]**2) < 1.0
  # cálculo estimado de PI.
  pi = 4 * (result.sum() /  n)
  # gráfico de pontos dentro e fora do cículo.
  if plot:
    fig, ax = plt.subplots(figsize=(7, 7))
    ax.plot(points[result, 0], points[result, 1], 'o', color='crimson')
    ax.plot(points[~result, 0], points[~result, 1], 'o', color='steelblue')
    ax.axis('equal')
    ax.axis([0, 1, 0, 1])
  return pi
```

```python
pi(10**3, plot=True)
```

```python
pi(10**4, plot=True)
```

```python
pi(10**5, plot=True)
```

```python
pi(10**6, plot=True)
```

```python
# Cálculo de PI a cada iteração.
for i in range(9):
    print(10**i, pi(10**i))
```
